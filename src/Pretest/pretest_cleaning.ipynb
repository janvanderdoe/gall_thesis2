```python
readability = data.iloc[2:,22:37]
readability.columns
col_names = []
for i in readability.columns:
    i = i.replace("Q Des ", "")
    col = i.split('_')
    col_name = "Q"+col[1]+"D"+col[0]
    col_names.append(col_name)
readability.columns = col_names
    
```

```python
readability = readability.reset_index()
readability
```

```python
columns = ['index', 'Description', 'Question', 'Value']

# Create an empty list to store the transformed data
long_data = []

index_col = readability['index']
# Iterate over the columns of the original dataframe
for column in readability.columns[1:]:
    # Split the column name into description and question parts
    question, description = column.split('D')
    
    # Extract the values from the original dataframe column
    values = readability[column].values.tolist()
    
    # Create tuples of (description, question, value) for each row
    rows = list(zip(index_col, [description]*len(values), [question]*len(values), values))
    
    # Extend the long_data list with the rows
    long_data.extend(rows)

# Create a new dataframe from the long_data list using the defined columns
long_df = pd.DataFrame(long_data, columns=columns)
long_df
```

```python
data = data.reset_index()

full_data = pd.merge(data, long_df, on = "index")
```

```python
full_data = full_data[full_data.columns.drop(list(full_data.filter(regex=r'Q Des')))]
```

```python
unfamiliarities = []
number_of_words = []

for index, row in descriptions.iterrows():
    length = len(sent_tokenize(str(row['Text'])))
    familiarity = 0
    bag_of_words = row['Text'].split()
    for word in bag_of_words:
        if word.lower() in top_3000:
            familiarity += 1
        elif word.lower() in top_3000_lemma:
            familiarity += 1
    unfamiliarity = 1- familiarity / len(bag_of_words)
    print(unfamiliarity)
    number_words = len(bag_of_words)/length
    print(number_words)
    unfamiliarities.append(unfamiliarity)
    number_of_words.append(number_words)

descriptions['unfamiliarity'] = unfamiliarities
descriptions['number_words'] = number_of_words
descriptions
```
